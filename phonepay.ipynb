```python
import pandas as pd
import json
import os

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

clm={'States':[],'Years':[],'Quaters':[],'Transacion_type':[],'Transacion_count':[],'Transacion_amount':[]}
for i in Agg_state_list:
 p_i=path+i+"/"
 Agg_yr=os.listdir(p_i)
 for j in Agg_yr:
   p_j=p_i+j+"/"
   Agg_yr_list=os.listdir(p_j)
   for k in Agg_yr_list:
       p_k=p_j+k
       Data=open(p_k,'r')
       D=json.load(Data)
       for z in D['data']['transactionData']:
         Name=z['name']
         count=z['paymentInstruments'][0]['count']
         amount=z['paymentInstruments'][0]['amount']
         clm['Transacion_type'].append(Name)
         clm['Transacion_count'].append(count)
         clm['Transacion_amount'].append(amount)
         clm['States'].append(i)
         clm['Years'].append(j)
         clm['Quaters'].append(int(k.strip('.json')))
Agg_Trans=pd.DataFrame(clm)
Agg_Trans
```

```python
import pandas as pd
import json
import os

path="/content/pulse/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

clm={'States':[],'Years':[],'Quaters':[],'User_brand':[],'User_count':[],'User_percentage':[],}
for i in Agg_state_list:
 p_i=path+i+"/"
 Agg_yr=os.listdir(p_i)
 for j in Agg_yr:
   p_j=p_i+j+"/"
   Agg_yr_list=os.listdir(p_j)
   for k in Agg_yr_list:
       p_k=p_j+k
       Data=open(p_k,'r')
       D=json.load(Data)
       if D['data']['usersByDevice']:
        for z in D['data']['usersByDevice']:
          brand=z['brand']
          count=z['count']
          percentage=z['percentage']
          clm['User_brand'].append(brand)
          clm['User_count'].append(count)
          clm['User_percentage'].append(percentage)
          clm['States'].append(i)
          clm['Years'].append(j)
          clm['Quaters'].append(int(k.strip('.json')))
Agg_User=pd.DataFrame(clm)
Agg_User
```

```python
import pandas as pd
import json
import os

path="/content/pulse/data/map/transaction/hover/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

clm={'States':[],'Years':[],'Quaters':[],'Hover_amount':[],'Hover_count':[],'District_name':[]}
for i in Agg_state_list:
 p_i=path+i+"/"
 Agg_yr=os.listdir(p_i)
 for j in Agg_yr:
   p_j=p_i+j+"/"
   Agg_yr_list=os.listdir(p_j)
   for k in Agg_yr_list:
       p_k=p_j+k
       Data=open(p_k,'r')
       D=json.load(Data)
       if D['data']['hoverDataList']:
        for z in D['data']['hoverDataList']:
          Name=z['name']
          count=z['metric'][0]['count']
          amount=z['metric'][0]['amount']
          clm['District_name'].append(Name)
          clm['Hover_count'].append(count)
          clm['Hover_amount'].append(amount)
          clm['States'].append(i)
          clm['Years'].append(j)
          clm['Quaters'].append(int(k.strip('.json')))
Map_Trans=pd.DataFrame(clm)
Map_Trans
```

```python
import os
import json
import pandas as pd

root_dir='/content/pulse/data/map/user/hover/country/india/state'
data_list=[]

for state_dir in os.listdir(root_dir):
  state_path=os.path.join(root_dir,state_dir)
  if os.path.isdir(state_path):

        for year_dir in os.listdir(state_path):
            year_path=os.path.join(state_path,year_dir)
            if os.path.isdir(year_path):

               for json_file in os.listdir(year_path):
                   if json_file.endswith('.json'):
                       with open(os.path.join(year_path,json_file)) as f:
                            data=json.load(f)

                            for district,values in data['data']['hoverData'].items():
                                row_dict={
                                     'States':state_dir,
                                     'Years':year_dir,
                                     'Quarters':int(json_file.split('.')[0]),
                                     'District':district,
                                     'RegisteredUser':values['registeredUsers'],
                                 }
                            data_list.append(row_dict)


Map_User=pd.DataFrame(data_list)
Map_User



```

```python
import os
import json
import pandas as pd

root_dir='/content/pulse/data/top/transaction/country/india/state'
data_list=[]

for state_dir in os.listdir(root_dir):
  state_path=os.path.join(root_dir,state_dir)
  if os.path.isdir(state_path):

        for year_dir in os.listdir(state_path):
            year_path=os.path.join(state_path,year_dir)
            if os.path.isdir(year_path):

               for json_file in os.listdir(year_path):
                   if json_file.endswith('.json'):
                       with open(os.path.join(year_path,json_file)) as f:
                            data=json.load(f)

                            for districts in data['data']['districts']:
                                row_dict={
                                     'States':state_dir,
                                     'Years':year_dir,
                                     'Quarters':int(json_file.split('.')[0]),
                                     'District':districts['entityName'],
                                     'Transaction_Type':districts['metric']['type'],
                                     'Transaction_Count':districts['metric']['count'],
                                     'Transaction_Amount':districts['metric']['amount'],

                                 }
                            data_list.append(row_dict)


Top_Trans=pd.DataFrame(data_list)
Top_Trans
```
